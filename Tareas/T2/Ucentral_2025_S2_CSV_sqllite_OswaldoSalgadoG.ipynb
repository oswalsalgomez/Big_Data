#5.1 Llenado de Tablas Maestras

```python
def cargar_datos_a_tablas_maestras(df, columna, tabla_destino):
    valores_unicos=df[columna].unique()
    for valor in valores_unicos:
        if(pd.notna(valor)):#ingnora valores NaN
            try:
                cursos.execute(f"INSERT OR IGNORE INTO {tabla_destino} (Nombre_{tabla_destino.replace('_','')}) VALUES (?)", (valor,))
            except sqlite3.Error as e:
                print(f"Error al insertar el tabla {tabla_destino}: {e}")
```

```python
def revisar_contenido_de_una_tabla(tabla_nombre,whereColumna='',whereValor='', limit=10, order_by_columna='', order_asc=True):
    try:
        conn=sqlite3.connect(db_path)
        cursos=conn.cursor()
        cursos.execute("SELECT COUNT(*) FROM "+tabla_nombre)
        total_registros=cursos.fetchone()[0]
        print(f"Total de registros en la tabla {tabla_nombre}: {total_registros}")
        sql="SELECT * FROM "+tabla_nombre
        if len(whereColumna)>0:
            sql+=" WHERE "+whereColumna+"=?"
        if (order_by_columna):
            sql += " ORDER BY " + order_by_columna + (" ASC" if order_asc else " DESC")
        if limit > 0:
            sql += " LIMIT " + str(limit)

        if len(whereColumna) > 0:
            cursos.execute(sql, (whereValor,))
        else:
            cursos.execute(sql)

        resultados = cursos.fetchall()
        for fila in resultados:
            print(fila)

    except sqlite3.Error as e:
        print(f"Error al revisar la tabla {tabla_nombre}: {e}")
    finally:
        if conn:
            conn.close()
```

#5.3 Tabla "Hija" Municipio

```python
cursos.execute('''
CREATE TABLE IF NOT EXISTS Municipio (
    Id_Municipio INTEGER PRIMARY KEY AUTOINCREMENT,
    Nombre_Municipio TEXT UNIQUE NOT NULL,
    Id_Departamento INTEGER,
    FOREIGN KEY (Id_Departamento) REFERENCES Departamento (Id_Departamento)
    UNIQUE (Nombre_Municipio, Id_Departamento)
)
''')
```

```python
# Para cada municipio buscar su departamento
def cargar_municipio(df, db_path):
    tot_creados = 0
    conn = None  # Initialize conn to None
    try:
        conn = sqlite3.connect(db_path)
        cursos = conn.cursor()
        for index, row in df.iterrows():
            departamento = row["Departamento"]
            municipio = row["Municipio"]
            cursos.execute("SELECT Id_Departamento FROM Departamento WHERE Nombre_Departamento=?", (departamento,))
            departamento_id = cursos.fetchone()
            if departamento_id:
                departamento_id = departamento_id[0]
                try:
                    cursos.execute("INSERT OR IGNORE INTO Municipio (Nombre_Municipio, Id_Departamento) VALUES (?,?)", (municipio, departamento_id))
                    tot_creados += 1
                except sqlite3.IntegrityError:
                    pass  # Ignore duplicate entries
            else:
                print(f"ERROR: Departamento {departamento} para el municipio {municipio} no encontrado")
        conn.commit()
    except sqlite3.Error as e:
        print(f"Database error: {e}")
    finally:
        if conn:
            conn.close()
    return tot_creados
```

6. Integrar con Tutelas Radicadas

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
url_folder_csv="/content/drive/MyDrive/3. Semestre 3/Big Data/Talleres/T_2"
columnas_nombre=["clasificacion",
                 "num_expediente",
                 "fecha_radica",
                 "demandante",
                 "demandado",
                 "primera_instancia",
                 "segunda_instancia",
                 "num_23Digitos",
                 "providencia",
                 "fecha_setencia"]
df_tutelas=pd.DataFrame(columns=columnas_nombre)
display(df_tutelas)
```

```python
def cargar_data_desde_archivo_csv(ruta_archivo, archivo_nombre):
    try:
        df_temporal=pd.read_csv(ruta_archivo,sep=";",header=None, encoding="latin-1", on_bad_lines="skip")
        #Agregar
        df_temporal.columns=columnas_nombre
        if (len(df_temporal.columns)==len(columnas_nombre)):
            print(f" archivo{archivo_nombre} carga exitosamente")
            return df_temporal
        else:
            print(f"archivo no trabajado {archivo_nombre} no tiene 10 columnas")
            return None

    except Exception as e:
        print(f"Error al procesar el archivo {ruta_archivo}: {str(e)}")
        return None
```

```python
sd_list=[] #lista de df resultante de los datos de cada uno de los csv
listado_archivos = [f for f in os.listdir(url_folder_csv) if f.endswith('.csv')]
with ThreadPoolExecutor() as executor:
    future_to_file={executor.submit(cargar_data_desde_archivo_csv, os.path.join(url_folder_csv, archivo_nombre), archivo_nombre): archivo_nombre for archivo_nombre in listado_archivos}
    for future in as_completed(future_to_file):
        df=future.result()
        if df is not None:
            sd_list.append(df)

# Concatenate all dataframes in the list
if sd_list:
    df_tutelas_parallel = pd.concat(sd_list, ignore_index=True)
    # Replace "NULL" with None
    df_tutelas_parallel.replace("NULL", None, inplace=True)
    display(df_tutelas_parallel.head())
```
